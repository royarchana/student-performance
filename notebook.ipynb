```python
import pandas as pd
import numpy as n
```

```python
df = pd.read_csv('Student_Performance.csv',header=0)
df.head()
```

```python
#handling missing value

df.isna().sum()
```

```python
df['Extracurricular Activities']=df['Extracurricular Activities'].replace({'Yes':1,"No":0})
```

```python
#after changing true to 1 and false to 0 ,the  changed data is given below
df.head()
```

```python
#
feature_names = df.columns
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#Feature selection

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler((0,1))

```

```python
x = scaler.fit_transform(x)
x
```

```python
x_df = pd.DataFrame(x,columns=feature_names[:-1])
x_df.head()
```

```python
#Splitting
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test = tts(x,y,train_size=0.75)
x_train.shape,y_train.shape
```

```python
#Model call for MLE

from sklearn.linear_model import LinearRegression


model = LinearRegression()


model.fit(x_train,y_train)
```

```python
#for model accuracy-use adjacent R2 Score


print("Model Accuracy: ", model.score(x_train, y_train))
```

```python
#

import pickle


pickle.dump(model,open('model.pkl','wb'))

```
